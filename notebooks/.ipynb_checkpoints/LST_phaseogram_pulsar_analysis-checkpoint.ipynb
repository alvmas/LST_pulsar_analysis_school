# 1. Creating the PulsarAnalysis object and settings

```python
h=PulsarAnalysis()
```

### 1.1 Setting the input file. 

For LST-1 data we use the function .setLSTInputFile() that has the following arguments:

    -- filename: single file to be read and used in the analysis
    -- dirname: directory where all the files that we include in the analysis are located
    -- src_dep: flag. Set True if your file/files have source-dependent format

```python
h.setLSTInputFile(dirname='/fefs/aswg/workspace/alvaro.mas/data/real/Crab/src_dependent/v0.9_crab/DL2_pulsar_performance/',src_dep=True)
#h.setLSTInputFile(dirname='/fefs/aswg/workspace/alvaro.mas/data/real/Crab/src_dependent/v0.9_2/DL2_total/')
#h.setLSTInputFile(filename='/fefs/aswg/workspace/alvaro.mas/data/real/Crab/src_dependent/v0.9_crab/dl2_LST-1_pulsar_nov20_feb22_gammaness05_alpha12.h5')
```

We can also as a general case give a list of phases, times and energies as an input using the function h.setListsInput():

    -- plist: list of phases
    -- tlist: list of times
    -- elist: list of energies

```python
#Example of use:
#h.setListsInput(plist=phase_list[0:int(4934143/2)],tlist=time_list[0:int(4934143/2)])
```

### 1.2. Setting the limits of the ON and OFF zones in the phaseogram

There are two functions for setting these limits. One for the background and one for the Peaks. In both cases the input lists with the edges could have dimension 2 (if only on region is associated with the bkg or the peak) or dimension 2n (where n is the number of regions of these zone)

```python
h.setBackgroundLimits([0.52,0.87])
h.setPeaklimits(P1_limits=[0,0.026,0.983,1],P2_limits=[0.377,0.422],P3_limits=None)
```

Check that P1 limits have dimension 4 because the peak is centered at 0 (which is the edge of the phaseogram)!

### 1.3 Setting the binning of the phaseogram and the time interval 

We can also set the binning that we are going to use for the construction of the lightcurve:

        -- nbins: number of bins in the phaseogram
        -- xmin: minimum phase of the interval (generally it should be zero)
        -- xmax: maximum phase of the phaseogram (generally it should be one)

```python
h.setBinning(nbins=50,xmin=0,xmax=1)
```

We can also define the time interval (in seconds!) in which we update the statistics. For instance, if we set tint=3600, the statistics will be calculated every hour of accumulated time of observation.

```python
h.setTimeInterval(tint=3600*50)
```

### 1.4 Setting of the cuts applied to the DL2 files

Additional fixed cuts in gammaness or arrival direction can be set using .setParamCuts():

        -- gammaness_cut: cut in gammaness to be applied (float for fixed or list for energy-dependent cuts)
        -- alpha_cut: cut in alpha to be applied (float for fixed or list for energy-dependent cuts)
        -- theta2_cut: cut in theta2 to be applied (float for fixed or list for energy-dependent cuts)
        -- zd_cut: cut in zd to be applied (float for fixed or list for energy-dependent cuts)
        -- energy_binnin_cut: energy binning for energy-dependent cuts (list or None)

Example of fixed cuts:

```python
h.setParamCuts(gammaness_cut=0.5,alpha_cut=12,zd_cut=35)
```

Example of energy-dependent cuts:

```python
h.setParamCuts(gammaness_cut=[0.3,0.5,0.7,0.8],alpha_cut=[15,12,10,8],zd_cut=35,energy_binning_cut=[0.02,0.05,0.1,0.5,100])
```

### 1.6 Setting of the fitting model

We can set the fitting and the model using h.setFittingParams() with the following arguments:

      -- binned: True if binned fitting or False for unbinned likelihood fitting
      -- model: model to be used

__Available models__:

- Single gaussian ('gaussian')
- Double gausssian ('dgaussian')
- Assymetric doble gaussian ('asym_dgaussian')
- Double lorentzian ('lorentzian')

```python
h.setFittingParams(model='dgaussian',binned=False)
```

### 1.7 Setting the energy binning for the energy-dependent analysis

Last, we could define (optionally) an energy binning to perform the analysis in different energy bins. For this, we use .setEnergybinning():

```python
h.setEnergybinning(list(np.logspace(math.log10(0.025),math.log10(0.5),4)))
```

# 2. Executing analysis

To execute the full analysis we should just call the function .run()

```python
h.run()
```

This function calls the different subclasses and perform the full analysis taking into account the settings that we gave before.
